```python
import requests
import re


def get_content(article_name):
    # Set the URL for the Wikipedia API
    url = "https://en.wikipedia.org/w/api.php"

    # Set the parameters for the API request
    params = {
        "action": "query",
        "format": "json",
        "prop": "extracts",
        "exsectionformat": "wiki",
        "titles": article_name,
        "exintro": True,
        "explaintext": True
    }

    # Send a GET request to the API and retrieve the JSON response
    response = requests.get(url, params=params)
    return response.json()

data = get_content("Ozone layer")
#print(data)


def merge_contents(data): 
    page_id = next(iter(data["query"]["pages"]))
    article_text = data["query"]["pages"][page_id]["extract"]
    
    # Remove HTML tags and replace special characters
    clean_text = re.sub(r'<[^>]+>', '', article_text)
    clean_text = re.sub(r'\n+', '\n', clean_text)
    clean_text = clean_text.replace('\xa0', ' ')
    clean_text = clean_text.strip()


# Remove all non-alphabetic characters
    clean_text = re.sub(r'[^a-zA-Z\s]+', ' ', clean_text)

    return clean_text
# Remove table of contents and other extraneous information
    clean_text = re.sub(r'\n?\[edit\]\n?', '\n', clean_text)
    clean_text = re.sub(r'^.*Contents\n\n', '', clean_text)
    clean_text = re.sub(r'^.*See also\n\n', '', clean_text)
    clean_text = re.sub(r'^.*References\n\n', '', clean_text)
    clean_text = re.sub(r'^.*Further reading\n\n', '', clean_text)
    clean_text = re.sub(r'^.*External links\n\n', '', clean_text)

content = merge_contents(data)
#print(content)

def tokenize(content):
        
    # Assuming 'content' contains the text you want to split into words
    merge_content = re.findall(r'\b\w+\b', content)
    return merge_content
    
collection = tokenize(content)
print(collection)

def lower_collection(collection):
    lowercase_collection = [word.lower() for word in collection]
    return lowercase_collection

    
lowercase_collection = lower_collection(collection)
#print(lowercase_collection)



# a function to count the number of occurrences of each word in the collection.
def count_frequency(lowercase_collection):
    count_frequency_dict={}
    count_already=[]
    for word in lowercase_collection:
        if word in count_already:
            count_frequency_dict[word]+=1
        else:
            count_frequency_dict[word]=1
            count_already.append(word)
    return count_frequency_dict
count_frequencies=count_frequency(lowercase_collection)
count_frequencies
            
list(count_frequencies.items())# converting the dict_items data structure into a simple liste of tuples 

#a function to print the n most frequent words along with their frenquency
def print_most_frequent(frequencies,n):
    valid_freqs=[freq for freq in frequencies if isinstance(freq, tuple) and len(freq)==2]
    sorted_dict=sorted(valid_freqs,key=lambda x: x[1],reverse=True)
    for i in range(n):
        print(f"{sorted_dict[i][0]:<10}   {sorted_dict[i][1]:<5}")
list_of_tuple=list(count_frequencies.items())
most_frequent=print_most_frequent(list_of_tuple,20)
most_frequent

# a function to Visualize the n most frequent words by plotting them in a histogram
import matplotlib.pyplot as plt
def plot_most_frequent(frequencies,n):
    valid_freqs=[freq for freq in frequencies if isinstance(freq, tuple) and len(freq)==2]# check and add inside a new list only the elements that are tuple of length 2 ex: (the,38)
    sorted_dict=sorted(valid_freqs,key=lambda x: x[1],reverse=True)# sorting the tuples of the new list base on their second element. That's why we have x[1] 
    words=[]
    freq=[]
    for i in range(n):
        words.append(sorted_dict[i][0])
        freq.append(sorted_dict[i][1])
    plt.figure(figsize=(15,10))
    plt.barh(words,freq)
    plt.gca().invert_yaxis()  # Invert the y-axis to display the graph in an inverted manner
    plt.ylabel("Frequency")
    plt.xlabel("Words")
    plt.show()

plot_most_frequent(list_of_tuple,20)

stop_words = [ "the", "a", "of", "to", "in", "about","that","is","and","s","by","was","from","on","its","other","has","or","it","as","with","at","no","5","nm","per"]

# a function 'remove_stop_words' that takes as parameters a list of tokens, it removes the empty words provided in this list and returns the new list resulting from this filtering.
def remove_stop_words(frequencies,stop_words):
    valid_freqs=[freq for freq in frequencies if isinstance(freq, tuple) and len(freq)==2]
    sorted_dict=sorted(valid_freqs,key=lambda x: x[1],reverse=True)
    filtered_collection=[x for x in sorted_dict if x[0] not in stop_words]#filtering the tuples of the liste base on if thier first element is inside de stop_word list. All the tuples that have they first element inside the stop_word list are removing
    return filtered_collection
filtered_collection=remove_stop_words(list_of_tuple,stop_words)
filtered_collection

#visualizing in a historgram the 25 most frequent words in the filtered collection.
plot_most_frequent(filtered_collection,20)
```
